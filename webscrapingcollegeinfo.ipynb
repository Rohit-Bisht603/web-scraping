```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get("https://collegedunia.com/btech/nirf-ranked-colleges", headers = headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
soup.prettify()
```

```python
n = []
l = []
r = []
rk = []

for i in range(39):
    n.append(soup.find_all('h3', class_="jsx-2257997354")[i].text.strip())
    l.append(soup.find_all('span', class_="jsx-2257997354 pr-1 location")[i].text.strip())
    r.append(soup.find_all('span', class_="jsx-2257997354 lr-key text-lg text-primary d-block font-weight-medium mb-1")[i].text.strip())
    try:
        rk.append(soup.find_all('span', class_="jsx-3528982912 rank-span no-break")[i].text.strip())
    except:
        rk.append(np.nan)
df = pd.DataFrame({'College name':n, 'Location':l, 'Rating':r,'NIRF Rank':rk})
df
```
